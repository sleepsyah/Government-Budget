```python
 #@title Normalizing Variables

#final normalized set is called df_norm
df_norm=df_final.copy()

expend_cols=['(E001) Total Expenditure','(E012) Total Salaries & Wages (Z00)',"(E024) Total Educ-Direct Exp",'(E058) Total Hospital-Dir Exp','(E074) Hous & Com-Direct Exp', '(E077) Libraries-Direct Exp', '(E087) Police Prot-Direct Exp', '(E090) Public Welf-Direct Exp']


for col in expend_cols:
  df_norm[col] = df_norm[col].str.replace('$', '').astype(float)



fix_cols=["Infant Mortality Rate","High School Graduation Rate","Violent Crime","Property Crime"]
df_norm[fix_cols] = df_norm[fix_cols].apply(
    pd.to_numeric, errors="coerce"
)

numeric_cols=[
    "High School Graduation Rate", "Uninsured","Percent in Poverty","Infant Mortality Rate", "Evidence-Based Reading and Writing", "Math","Total","RATIO", "Percent of Labor Force Unemployed in State", "Violent Crime", "Property Crime",
]



df_norm[numeric_cols]=(df_norm[numeric_cols]-df_norm[numeric_cols].min())/(df_norm[numeric_cols].max()-df_norm[numeric_cols].min())

df_norm["Education Index"]=0.75*df_norm["Score"] + + 0.25*df_norm["High School Graduation Rate"]

df_norm["Health Index"]=0.55*df_norm["RATIO"] + 0.35*(1-df_norm["Infant Mortality Rate"]) + 0.2*(1-df_norm["Uninsured"])

df_norm["Social Index"]= 1- (0.35*df_norm["Violent Crime"] + 0.1*df_norm["Property Crime"] + 0.2*df_norm["Percent of Labor Force Unemployed in State"] + 0.35*df_norm["Percent in Poverty"])

df_norm["Index"] = df_norm[["Social Index","Health Index","Education Index"]].mean(axis=1)

# df_norm["Index"] = (df_norm["Social Index"]*df_norm["Health Index"]*df_norm["Education Index"])**(1/3)


print(df_norm.shape)
print(df_health[df_health["State"]=="Texas"])
```
